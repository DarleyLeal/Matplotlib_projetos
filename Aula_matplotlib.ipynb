1ª Aula sobre Matplotlib


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 11)
y = x ** 2

plt.title('Função')
plt.xlabel('Eixo X da Função')
plt.ylabel('Eixo Y da Função')

plt.plot(x, y, 'blue')
```

CRIANDO DOIS GRÁFICOS

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
```

ORIENTAÇÃO À OBJETOS COM MATPLTLIB

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #esquerda, inferior, largura, altura
axes.plot(x, y, 'blue')
axes.set_xlabel('Eixo de X')
axes.set_ylabel('Eixo de Y')
axes.set_title('Figura com eixos')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #esquerda, inferior, largura, altura
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) #esquerda, inferior, largura, altura

axes1.set_xlabel('Eixo de X1')
axes1.set_ylabel('Eixo de Y1')
axes1.set_title('Figura 1')
axes1.plot(x, y, 'blue')

axes2.set_xlabel('Eixo de X2')
axes2.set_ylabel('Eixo de Y2')
axes2.set_title('Figura 2')
axes2.plot(y, x, 'red')
```

SUBPLOTS

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8, 5))

for axis in ax:
  axis.plot(x, y, 'r')

plt.tight_layout()
```

LEGENDAS, TÍTULOS E RÓTULOS

```python
import numpy as np
import matplotlib.pyplot as plt


fig = plt.figure()

ax = fig.add_axes([0, 0 , 1, 1])


#label='x**2' é adicionado texto debtro da função
ax.plot(x, x **2, label='x**2', color='#FF50AB', alpha=1)
ax.plot(x, x **3, label='x**3', color="green", alpha=1)
#Método alpha adiciona transparências às linhas do eixo

ax.set_title("Gráfico da Função")
ax.set_xlabel("Eixo de X")
ax.set_ylabel("Eixo de Y")


ax.legend(loc=6)
```

MAIS EXEMPLOS

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# Possiveis estilos de linha: ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# Traços estilizados
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # Formato: comprimento da linha, comprimento do espaço, ...

# possíveis símbolos de marcador: marcador = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# tamanho e cor do marcador
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df_flights_matrix = flights.pivot_table(index='year', columns='month', values='passengers')
fig, ax = plt.subplots(figsize=(20, 8))

sns.heatmap(df_flights_matrix, ax=ax, annot=True, fmt=".0f")
#Caso eu queira separar os dados posso usar depois do fmt o linewdths=1 ou outro tamanho quaisquer
#USAMOS MAPA DE CALOR QUANDO TEMOS DUAS VÁRIAVEIS E QUEREMOS ANALISAR UMA TERCEIRA VÁRIAVEL CONTINUA
#POSSO MUDAR A PALETA DE CORES USANDO 'cmap=magma' NO CASO O MAGMA É UMA DAS OPÇÕES DISPONIVEIS DE CORES
```

```python
sns.clustermap(df_flights_matrix)
#REAGRUPA VALORES RELACIONADOS
```
