USANDO O MATPLOTLIB COM PYPLOT

```python
#PLOTAGEM DE DADOS
#Histograma

import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_csv('modificado.csv')

atributo = 'Defesa'
bins = range(0, 201, 20)
plt.xticks(bins)
plt.title(f'Distribuição de Pokemons por {atributo}')
plt.xlabel(f'Valor do atributo {atributo}')
plt.ylabel(f'Quantidade de Pokemons')
plt.hist(df[atributo], bins=bins)
plt.show()
```

GRÁFICO VERTICAL COM O MATPLOTLIB

```python
atributos = 'Velocidade'
tipos = df.groupby('Tipo 1').mean().sort_values(atributo, ascending=False)
linha = tipos.index.tolist()
coluna = tipos[atributo].tolist()
figura = plt.figure(figsize=(14, 4)).add_axes([0,0,1,1])
figura.bar(linha, coluna)

plt.xlabel('Tipos de Pokemon')
plt.ylabel(f'Valor médio do atributo {atributo}')
plt.title(f'Índices de Pokemons por {atributo}')

for i, v in enumerate(coluna):
  plt.annotate(f'{v:.2f}', (i-0.3, v + 1))

plt.show()
```
